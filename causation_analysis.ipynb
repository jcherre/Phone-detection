```python
import pandas as pd

df_ori = pd.read_csv('dummy_data.csv')
df_ori
```

```python
#Preprocess
df = df_ori.drop(0)
df = df.rename(columns={'periodo_creacion_seguimiento_comercial': 'date', 'periodo': 'periodo_le', 'periodo_1': 'periodo_pa'})
df['periodo_le']    = df['periodo_le'].fillna(df['date'])
df['periodo_pa']    = df['periodo_pa'].fillna(df['date'])
df['evento']        = df['evento'].fillna('Compra de Vehículo')
df['leads_cv']      = df['leads_cv'].fillna(0)
df['recuento']      = df['recuento'].fillna(0)
df
```

```python
from causalimpact import CausalImpact

# Example data: Simulated time-series data
#data = pd.DataFrame({
#    'date': pd.date_range(start='2023-01-01', periods=200),
#    'sales': range(100, 300),  # Simulated sales data
#    'marketing_spend': [0] * 100 + [1] * 100  # Intervention starts at day 100
#})
data = df[['date','mnt_oportunidad','leads_cv']]
data['date'] = pd.to_datetime(data['date'])

# Set date as the index
data.set_index('date', inplace=True)

# Define pre- and post-intervention periods
pre_period = ['2023-03-01', '2024-01-01']  # First 100 days
post_period = ['2024-02-01', '2025-02-01']  # Next 100 days

# Run causal impact analysis
impact = CausalImpact(data, pre_period, post_period)

# Print summary and plot results
print(impact.summary())
impact.plot()
```

```python
import dowhy
from dowhy import CausalModel

# Example data: Simulated dataset
#data = pd.DataFrame({
#    'treatment': [0, 1, 0, 1, 0, 1],  # Treatment variable
#    'outcome': [10, 20, 15, 25, 12, 22],  # Outcome variable
#    'confounder': [5, 10, 7, 12, 6, 11]  # Confounding variable
#})
data = df[['leads_cv','mnt_oportunidad','recuento']]
data.columns = ['treatment', 'outcome', 'confounder']

# Step 1: Create a causal model
model = CausalModel(
    data=data,
    treatment='treatment',
    outcome='confounder'
)

# Step 2: Identify the causal effect
#identified_estimand = model.identify_effect()

# Step 2: Identify causal effect and return target estimands
identified_estimand = model.identify_effect()

# Step 3: Estimate the target estimand using a statistical method.
estimate = model.estimate_effect(identified_estimand,
                                 method_name="backdoor.linear_regression")

# Step 4: Refute the obtained estimate using multiple robustness checks.
refute_results = model.refute_estimate(identified_estimand, estimate,
                                       method_name="random_common_cause")
print(estimate)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from causalinference import CausalModel

data = df[['leads_cv','mnt_oportunidad','recuento']]
data.columns = ['Treatment', 'Outcome', 'Confounder']

model = CausalModel(
    Y=data['Outcome'].values,
    D=data['Treatment'].values,
    X=data['Confounder'].values
)

model.est_via_ols()
model.est_via_matching()

# Visualizing the outcomes
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data[data['Treatment'] == 0]['Outcome'], alpha=0.5, label='Control', color='blue')
plt.hist(data[data['Treatment'] != 0]['Outcome'], alpha=0.5, label='Treated', color='red')
plt.title('Distribution of Outcomes')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 2, 2)
treated_mean = data[data['Treatment'] != 0]['Outcome'].mean()
control_mean = data[data['Treatment'] == 0]['Outcome'].mean()
plt.bar(['Control', 'Treated'], [control_mean, treated_mean], color=['blue', 'red'])
plt.title('Average Outcome by Group')
plt.ylabel('Average Outcome')

plt.tight_layout()
plt.show()
```
